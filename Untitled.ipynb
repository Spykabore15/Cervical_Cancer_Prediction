La suppression de ces colonnes ne semblera donc pas influencer nos prédictions. Pour les autres valeurs abbérentes, on les remplacerait par la médiane( pour les colonnes numériques) et le mode (pour les colonnes catégorielles) de leurs colonnes correspondantes. 

```python
# Traaitement des valeurs abbérentes
def treat_Nan_values(df, column):
    # Vérifier si la colonne est numérique
    if pd.api.types.is_numeric_dtype(df[column]):
        # Remplacer les valeurs manquantes par la médiane
        df[column] = df[column].fillna(df[column].median())
    else:
        # Pour les colonnes catégorielles, remplacer par le mode
        df[column] = df[column].fillna(df[column].mode()[0])
    
    return df

# Appliquer à toutes les colonnes numériques

for column in df.columns :
    df =  treat_Nan_values(df, column)

# Vérifier les résultats
print(df.isnull().sum())  # Doit retourner 0 pour toutes les colonnes traitées
```

 Code de Détection et gestion des outliers

```python
# Fonction pour détecter les outliers avec l'IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25e percentile
    Q3 = df[column].quantile(0.75)  # 75e percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Fonction pour traiter les outliers dans toutes les colonnes numériques
def treat_outliers_iqr(df):
    # Liste des colonnes numériques
    numeric_columns = df.select_dtypes(include=["int64", "float64"]).columns
    
    for column in numeric_columns:
        # Détecter les outliers
        outliers, lower_bound, upper_bound = detect_outliers_iqr(df, column)
        
        # Afficher les informations sur les outliers
        print(f"Colonne : {column}")
        print(f"Nombre d'outliers détectés : {len(outliers)}")
        print(f"Valeurs limites : [{lower_bound}, {upper_bound}]")
        print("------------------------")
        
        # Remplacer les outliers par les valeurs limites
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    
    return df

# Appliquer la fonction à ton DataFrame
df = treat_outliers_iqr(df)

# Vérifier les résultats avec un boxplot pour chaque colonne
for column in df.select_dtypes(include=["int64", "float64"]).columns:
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot de {column} après gestion des outliers")
    plt.show()
```

Les résultas ci dessous nous montre que les outliers ont tous été bien traités.
