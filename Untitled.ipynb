La suppression de ces colonnes ne semblera donc pas influencer nos prÃ©dictions. Pour les autres valeurs abbÃ©rentes, on les remplacerait par la mÃ©diane( pour les colonnes numÃ©riques) et le mode (pour les colonnes catÃ©gorielles) de leurs colonnes correspondantes. 

```python
# Traaitement des valeurs abbÃ©rentes
def treat_Nan_values(df, column):
    # VÃ©rifier si la colonne est numÃ©rique
    if pd.api.types.is_numeric_dtype(df[column]):
        # Remplacer les valeurs manquantes par la mÃ©diane
        df[column] = df[column].fillna(df[column].median())
    else:
        # Pour les colonnes catÃ©gorielles, remplacer par le mode
        df[column] = df[column].fillna(df[column].mode()[0])
    
    return df

# Appliquer Ã  toutes les colonnes numÃ©riques

for column in df.columns :
    df =  treat_Nan_values(df, column)

# VÃ©rifier les rÃ©sultats
print(df.isnull().sum())  # Doit retourner 0 pour toutes les colonnes traitÃ©es
```

 Code de DÃ©tection et gestion des outliers

```python
# Fonction pour dÃ©tecter les outliers avec l'IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25e percentile
    Q3 = df[column].quantile(0.75)  # 75e percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Fonction pour traiter les outliers dans toutes les colonnes numÃ©riques
def treat_outliers_iqr(df):
    # Liste des colonnes numÃ©riques
    numeric_columns = df.select_dtypes(include=["int64", "float64"]).columns
    
    for column in numeric_columns:
        # DÃ©tecter les outliers
        outliers, lower_bound, upper_bound = detect_outliers_iqr(df, column)
        
        # Afficher les informations sur les outliers
        print(f"Colonne : {column}")
        print(f"Nombre d'outliers dÃ©tectÃ©s : {len(outliers)}")
        print(f"Valeurs limites : [{lower_bound}, {upper_bound}]")
        print("------------------------")
        
        # Remplacer les outliers par les valeurs limites
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    
    return df

# Appliquer la fonction Ã  ton DataFrame
df = treat_outliers_iqr(df)

# VÃ©rifier les rÃ©sultats avec un boxplot pour chaque colonne
for column in df.select_dtypes(include=["int64", "float64"]).columns:
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot de {column} aprÃ¨s gestion des outliers")
    plt.show()
```

Les rÃ©sultas ci dessous nous montre que les outliers ont tous Ã©tÃ© bien traitÃ©s.
